# Datatypes in Pandas Using pandas.Dataframe select_dtypes method

```python
import pandas as pd

def create_records():

    student_data = {
        "RollNo": [201, 202, 203, 204, 205],
        "StudentName": ["Akhila", "srisha", "sravani", "dev", "Nikhil"],
        "Department": ["CSE", "Bcom.computers", "BCA", "IT", "BSc.computers"],
        "CGPA": [8.5, 9.1, 7.8, 8.9, 8.0],
        "Attendance": [92, 88, 75, 95, 81],
        "Hosteller": [True, False, True, True, False]
    }

    df = pd.DataFrame(student_data)

    df["StudentName"] = df["StudentName"].astype("string")
    df["Department"] = df["Department"].astype("string")

    return df
```

```python
def main():

    df = create_records()

    print("\nComplete DataFrame:\n")
    print(df)

    print("\nColumn Data Types:\n")
    print(df.dtypes)

    print("\nNumeric Data:\n")
    print(df.select_dtypes(include=["number"]))

    print("\nString Data:\n")
    print(df.select_dtypes(include=["string"]))

    print("\nBoolean Data:\n")
    print(df.select_dtypes(include=["bool"]))

    print("\nNon-Numeric Data:\n")
    print(df.select_dtypes(exclude=["number"]))

main()
```
